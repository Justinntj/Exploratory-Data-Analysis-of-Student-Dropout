```python
plt.figure(figsize=(6,4))
sns.histplot(df['unemployment rate'], discrete=True) 
plt.title('Unemployment Rate Distribution')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(df['inflation rate'])
plt.title('KDE Plot of inflation rate')
plt.xlabel('inflation rate')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# Shows the count of students for each marital status, split by graduation target
sns.countplot(x='marital status', hue='target', data=df, palette='Set2')

plt.title('Student Outcomes by Marital Status')
plt.xlabel('Marital Status Code')
plt.ylabel('Number of Students')
plt.legend(title='Outcome')
plt.show()
```
